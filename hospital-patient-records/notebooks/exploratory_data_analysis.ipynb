```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the processed data
data = pd.read_csv('../data/processed/processed_data.csv')

# Display the first few rows of the dataset
data
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
data.dtypes
```

```python
data['START'] = pd.to_datetime(data['START'])
data['STOP'] = pd.to_datetime(data['STOP'])
data['BIRTHDATE'] = pd.to_datetime(data['BIRTHDATE'])
data['DEATHDATE'] = pd.to_datetime(data['DEATHDATE'])
data['ZIP'] = data['ZIP'].astype(str)
data['ZIP_ORG'] = data['ZIP_ORG'].astype(str)
data['START_PROCEDURE'] = pd.to_datetime(data['START_PROCEDURE'])
data['STOP_PROCEDURE'] = pd.to_datetime(data['STOP_PROCEDURE'])
```

```python
# Summary statistics of numerical columns
data.describe()
```

```python
# Plot the number of admissions and readmissions over time
data['ADMISSION_DATE'] = pd.to_datetime(data['START']).dt.date
admissions_over_time = data.groupby('ADMISSION_DATE').size()

plt.figure(figsize=(10, 6))
admissions_over_time.plot()
plt.title('Number of Admissions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.show()

# Plot the number of readmissions over time
readmissions_over_time = data[data['READMISSION'] == 1].groupby('ADMISSION_DATE').size()

plt.figure(figsize=(10, 6))
readmissions_over_time.plot()
plt.title('Number of Readmissions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Readmissions')
plt.show()
```

```python
# Calculate and plot the average length of stay over time
average_length_of_stay = data.groupby('ADMISSION_DATE')['DURATION'].mean()

plt.figure(figsize=(10, 6))
average_length_of_stay.plot()
plt.title('Average Length of Stay Over Time')
plt.xlabel('Date')
plt.ylabel('Average Length of Stay (hours)')
plt.show()
```

```python
# Plot the number of readmissions by gender
readmissions_by_gender = data[data['READMISSION'] == 1].groupby('GENDER')['READMISSION'].count()

plt.figure(figsize=(10, 6))
readmissions_by_gender.plot(kind='bar')
plt.title('Number of Readmissions by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Readmissions')
plt.show()
```

```python
# Plot the number of readmissions by race
readmissions_by_race = data[data['READMISSION'] == 1].groupby('RACE')['READMISSION'].count()

plt.figure(figsize=(10, 6))
readmissions_by_race.plot(kind='bar')
plt.title('Number of Readmissions by Race')
plt.xlabel('Race')
plt.ylabel('Number of Readmissions')
plt.show()
```

```python
# Calculate and plot the average cost per visit
average_cost_per_visit = data.groupby('ADMISSION_DATE')['TOTAL_CLAIM_COST'].mean()

plt.figure(figsize=(10, 6))
average_cost_per_visit.plot()
plt.title('Average Cost Per Visit Over Time')
plt.xlabel('Date')
plt.ylabel('Average Cost ($)')
plt.show()

# Distribution of costs
plt.figure(figsize=(10, 6))
sns.histplot(data['TOTAL_CLAIM_COST'], bins=50, kde=True)
plt.title('Distribution of Total Claim Costs')
plt.xlabel('Total Claim Cost ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate and plot the percentage of procedures covered by insurance
data['COVERED_BY_INSURANCE'] = data['PAYER_COVERAGE'] > 0
insurance_coverage = data.groupby('ADMISSION_DATE')['COVERED_BY_INSURANCE'].mean() * 100

plt.figure(figsize=(10, 6))
insurance_coverage.plot()
plt.title('Percentage of Procedures Covered by Insurance Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage Covered by Insurance (%)')
plt.show()

# Analyze the variation in coverage across different procedures
procedure_coverage = data.groupby('DESCRIPTION')['COVERED_BY_INSURANCE'].mean() * 100

plt.figure(figsize=(10, 6))
procedure_coverage.sort_values().plot(kind='barh')
plt.title('Insurance Coverage by Procedure')
plt.xlabel('Percentage Covered by Insurance (%)')
plt.ylabel('Procedure')
plt.show()
```

```python
# Plot the distribution of durations
plt.figure(figsize=(10, 6))
sns.histplot(data['DURATION'], bins=50, kde=True)
plt.title('Distribution of Admission Durations')
plt.xlabel('Admission Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of insurance coverage
plt.figure(figsize=(10, 6))
sns.histplot(data['INSURANCE_COVERAGE'], bins=50, kde=True)
plt.title('Distribution of Insurance Coverage')
plt.xlabel('Insurance Coverage')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of total claim costs
plt.figure(figsize=(10, 6))
sns.histplot(data['TOTAL_CLAIM_COST'], bins=50, kde=True)
plt.title('Distribution of Total Claim Costs')
plt.xlabel('Total Claim Cost ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of base encounter costs
plt.figure(figsize=(10, 6))
sns.histplot(data['BASE_ENCOUNTER_COST'], bins=50, kde=True)
plt.title('Distribution of Base Encounter Costs')
plt.xlabel('Base Encounter Cost ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(data[['BASE_ENCOUNTER_COST', 'TOTAL_CLAIM_COST']]))
outliers = np.where(z_scores > 3)

plt.figure(figsize=(10, 6))
plt.scatter(data['BASE_ENCOUNTER_COST'], data['TOTAL_CLAIM_COST'], label='Data')
plt.scatter(data.iloc[outliers[0]]['BASE_ENCOUNTER_COST'], data.iloc[outliers[0]]['TOTAL_CLAIM_COST'], color='red', label='Outliers')
plt.xlabel('Base Encounter Cost')
plt.ylabel('Total Claim Cost')
plt.title('Scatter Plot with Outliers')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['LON'], data['LAT'])
plt.title('Scatter Plot of Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['START'].value_counts().sort_index().plot(kind='line')
plt.title('Line Plot of Start Dates')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['GENDER'].value_counts().plot(kind='bar')
plt.title('Bar Plot of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['AGE_AT_ENCOUNTER'].hist(bins=50)
plt.title('Histogram of Age at Encounter')
plt.xlabel('Age at Encounter')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.boxplot(column=['AGE_AT_ENCOUNTER', 'INSURANCE_COVERAGE', 'DURATION'])
plt.title('Box Plot of Age, Insurance Coverage, and Duration')
plt.show()
```

```python
# input outliers in the duration column
data['DURATION'] = data['DURATION'].clip(lower=0, upper=data['DURATION'].quantile(0.99))
```

```python
data['ADMISSION_DATE'] = pd.to_datetime(data['ADMISSION_DATE'])
```

```python
# Save the data
data.to_csv('../data/processed/records.csv', index=False)
```
